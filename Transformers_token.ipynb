```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
```

```python
classifier('The food is not that great but crust is awesome')
```

```python
results = classifier(['we are happy to show the Transfomers libray.',"we hopr you don't hate it."])
for result in results:
  print(f"label: {result['label']}, with score: {round(result['score'],4)}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
model = TFAutoModelForSequenceClassification.from_pretrained(model_name,from_pt=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline('sentiment-anlaysis', model=model , tokenizer=tokenizer)
```

```python
classifier ("I am good boy")
```
